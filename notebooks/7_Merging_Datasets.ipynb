```python
data_bitcoin = pd.read_csv('bitcoin_prices.csv')
```

```python
data_bitcoin.head()
```

```python
data_bitcoin.columns = ['date', 'open', 'high', 'low', 'close', 'adj_close', 'volume']
```

```python
data_bitcoin['date'] = pd.to_datetime(data_bitcoin['date'], format='%Y-%m-%d').dt.floor('D')
```

```python
year = 2021
from_month = 2
to_month = 8
data_bitcoin_fil= data_bitcoin.loc[(data_bitcoin['date'].dt.year == year) & ((data_bitcoin['date'].dt.month >= from_month) & (data_bitcoin['date'].dt.month <= to_month))]
```

###############################################

```python
tweet_roberta_score = pd.read_csv("roberta_aggregated.csv")
```

```python
tweet_roberta_score['date'] = pd.to_datetime(tweet_roberta_score['date'], format='%Y-%m-%d').dt.floor('D')
```

```python
tweet_roberta_score.dtypes
```

```python
data_r = pd.merge(data_bitcoin_fil, tweet_roberta_score, on="date", how="outer")
```

```python
data_r = data_r.fillna(0)
```

```python
data_r.head()
```

```python
data_final_r = data_r[['date', 'close', 'compound_score']]
```

```python
data_final_r.head()
```

```python
%cd /content/gdrive/MyDrive/FinalThesis/Final_Data/
```

```python
data_final_r.to_csv("roberta_bitcoin_merged.csv", index=False)
```
